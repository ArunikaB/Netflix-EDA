# CLEANING DATASET

```python
#filling missing data

df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
#droping missing data 

df= df.dropna()
```

```python
#checking for NaN
df.isna().sum()
```

# FIXING DATE COLUMN

```python
df = df.copy()

df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip())
df['month_added']=df['date_added'].dt.month
df['month_name_added']=df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year

df
```

# DATA VISUALIZATION

**1. Movie to TV Show ratio**

```python
plt.figure()
type_counts = df['type'].value_counts()

plt.pie(type_counts, labels=type_counts.index, colors=['red','maroon'], autopct='%1.1f%%')
plt.title('Percentage Distribution')
plt.show()
```

**2. Top content producing countries**

```python
top_10 = df.country.value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_10.index, x=top_10.values, palette='rocket')
plt.title('Top 10 Countries by Netflix Content Production')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

**3. Most common genres**

```python
genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in genres for genre in sublist]

genre_series = pd.Series(flat_genres)
top_genres = genre_series.value_counts().head(10)

genres_df = top_genres.reset_index()
genres_df.columns = ['Genre', 'Count']

plt.figure(figsize=(10, 5))
plt.barh(genres_df['Genre'], genres_df['Count'], color='darkred')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

**4. Releases over time**

```python
time_series = df.groupby(['release_year', 'type']).size().unstack()
plt.figure(figsize=(12,6))
sns.lineplot(data=time_series, dashes=False, markers=True)
plt.title('Netflix Content Production Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.legend(title='Content Type')
plt.grid(True)
plt.show()

most= time_series.sum(axis=1).idxmax()
print("\nYear with the maximum releases: ", most)
```
